# 03. Regional Data

```python
ords_prods_cust['STATE'].value_counts(dropna = False)
```

```python
region = []

for value in ords_prods_cust['STATE']:
    if value in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']:
        region.append('Northeast')
    elif value in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']:
        region.append('Midwest')
    elif value in ['Delaware','Maryland','District of Columbia','Virginia','West Virginia','North Carolina','South Carolina','Georgia','Florida','Kentucky','Tennessee','Mississippi','Alabama','Oklahoma','Texas','Arkansas','Louisiana']:
        region.append('South')
    else:
        region.append('West')
```

```python
region
```

```python
ords_prods_cust['region'] = region
```

```python
# Check new column
ords_prods_cust['region'].value_counts(dropna = False)
```

```python
ords_prods_cust.head()
```

```python
# Create crosstab between 'region' and 'spending_flag'
region_spending = pd.crosstab(ords_prods_cust['region'], ords_prods_cust['spending_flag'], dropna = False)
```

```python
region_spending
```

### There is little to no difference between the regions when comparing their high and low spenders.

# 04. Low-Activity Customers

```python
# Creating a low activity flag
ords_prods_cust.loc[ords_prods_cust['max_order'] < 5, 'low_activity_flag'] = 'low_activity'
```

```python
# Create a high activity flag
ords_prods_cust.loc[ords_prods_cust['max_order'] >= 5, 'low_activity_flag'] = 'high_activity'
```

```python
# Check new flag
ords_prods_cust['low_activity_flag'].value_counts(dropna = False)
```

```python
# Create a subset with no low-activity
cust_high_activity = ords_prods_cust[ords_prods_cust['low_activity_flag'] == 'high_activity']
```
