# 03. Create User-Defined Function

```python
# Define a User-Defined Function for the Three Price Ranges
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

# 04. Create If-Statements Using loc( ) Function

### Loc( ) Statements to filter the subset

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

### Loc( ) Statements to filter the entire dataframe

```python
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

# If-Statements with For-Loops for Dataframe

```python
# Finding which days of the week have the highest/lowest frequency
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
ords_prods_merge['busiest_day'] = result
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

# Creating Two Busiest/Slowest Days Label

```python
# Finding which days of the week have the highest/lowest frequency
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Creating new label for two busiest/slowest days
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 & 1:
    result.append("Busiest days")
  elif value == 4 & 3:
    result.append("Least busiest days")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
ords_prods_merge['busiest_days'] = result
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

# Creating Busiest Hours of Day Label

```python
ords_prods_merge.head()
```

```python
# Finding which hours have the highest/lowest frequency
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
result = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value == 10:
    result.append("Most orders")
  elif value == 3:
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
result
```

```python
ords_prods_merge['busiest_period_of_day'] = result
```

```python
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

### The amount of most orders matches up with the value of 10, the amount of fewest orders matches up with the value of 3, and the average orders matches with every other value. This means that our new label 'busiest_period_of_day' matches up with the values under 'order_hour_of_day'.
